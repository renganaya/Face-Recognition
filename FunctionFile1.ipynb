```python
#import libraries
import numpy as np
import cv2 
import os

```

```python
def FaceRec_Function1(img):
    
 
    #loading the haarcascade_frontalface xml file
    face_cascade = cv2.CascadeClassifier(r'C:\Users\NEW\haarcascade_frontalface_alt.xml')

    #converting original image to grayscale for the training
    img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)                   
    
    #detect multiscale images
    faces_rect = face_cascade.detectMultiScale(img_gray,1.3,7)
    
    return faces_rect,img_gray
```

```python
 #unit testing for Function1
    
img=cv2.imread(r'C:\FrProject\images\0\AnithaImg18.jpg')

faces_rect,img_gray=FaceRec_Function1(img)
    
while True:
 
    cv2.imshow('Gray Image',img_gray)
    if cv2.waitKey(1) == 13: #press Enter key
        break
    
cv2.destroyAllWindows()
```

```python
#parse through the image directory and subfolders for training
def FaceRec_Function2(img_directory):

    faces = []
    faceid = []

    for path,subdirnames,filenames in os.walk(img_directory):
        
        #parse filename in subfolder
         for filename in filenames:
            
            #test the systems file
            if filename.startswith("."):
                print("Skipped the System files")
                print("Skipped File:",filename)
                continue
           
            #trying to get the subdirectory name
            img_dirid=os.path.basename(path)
            print("FaceId:",img_dirid)
            
            
            #join the path and filename together
            img_fullpath=os.path.join(path,filename)
            print("Filename Path:",img_fullpath)
         
            test_img=cv2.imread(img_fullpath)
            
            #file has no image
            if test_img is None:
                print("No image found:", img_fullpath)
                
            #calling Function1
            faces_rect,gray_img=FaceRec_Function1(test_img)
           
            #if there's no faces detected,ignore and continue
            if len(faces_rect) == 0:
                
                print("No faces detected:", img_fullpath)
                continue
            
            (x,y,w,h) = faces_rect[0]
            crop_img = gray_img[y:y+w,x:x+h]
            
            faces.append(crop_img)
            faceid.append(int(img_dirid))
            
    print("Function2 finished")
    return faces,faceid
```

```python
#Train the faces using LBPH algorithm
def FaceRec_Function3(faces,faceid):
    
    #declare LBPH algorithm
    FaceRecognizer=cv2.face.LBPHFaceRecognizer_create() 
    
     #train the algorithm
    FaceRecognizer.train(faces,np.array(faceid))
    
    return FaceRecognizer
```

```python
Face_Rect,FaceId=FaceRec_Function2(r"C:\FrProject\images\0")
FaceRecognizer=FaceRec_Function3(Face_Rect,FaceId)
```

```python
#draw the rectanle on the detected face
def FaceRec_Function4(test_img,face):
    (x,y,w,h)=face
    #put the rectangle onto the test_img
    cv2.rectangle(test_img,(x,y),(x+w,y+h),(255,0,0),thickness=3)
```

```python
#put the text on the face rectangle 
def FaceRec_Function5(test_img,text,x,y):
#recieve the image,text the starting position of the image
    cv2.putText(test_img,text,(x,y),cv2.FONT_HERSHEY_COMPLEX,0.5,(0,0,255),2)
```
