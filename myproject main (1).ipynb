```python


import cv2
import os
import FunctionFile1 as fn





```

```python
test_img=cv2.imread(r'C:\FrProject\images\1\sharukImg307.jpg')
         
faces_detected,gray_img=fn.FaceRec_Function1(test_img)
print("Face Detected:",faces_detected)
```

```python
faces,faceid=fn.FaceRec_Function2(r"C:\FrProject\images\1")


face_recognizer=fn.FaceRec_Function3(faces,faceid)

                              
face_recognizer.write("training.yml")         
```

```python
name={0:"Anitha",1:"Sharuk Khan",2:"Dhayanithi"}

for face in faces_detected:
    (x,y,w,h)=face
    roi_gray=gray_img[y:y+w,x:x+h]
    label,confidence=face_recognizer.predict(roi_gray)
    
    print("Label:",label)
    print("Confidence:",confidence)
    
    fn.FaceRec_Function4(test_img,face)
    predict_name=name[label]
    
    #Testing-------------
    print("predicted Name:",predict_name)
    #Testing-------------
    
    #checking the confidence level
    if confidence > 105:
        continue
    else:
        fn.FaceRec_Function5(test_img,predict_name,x,y)
        resizedimage=cv2.resize(test_img,(700,600))
        
        #testing output------
        while True:
            cv2.imshow("Cropped Final Image",resizedimage)
            if cv2.waitKey(1) == 13: #Enter key
    
                break
        cv2.destroyAllWindows()
    
```
