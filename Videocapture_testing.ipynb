```python

import cv2
import FunctionFile1 as fn


face_recognition=cv2.face.LBPHFaceRecognizer_create()
face_recognition.read("training.yml")
```

```python
def FaceRec_Function4(test_img,face):
    (x,y,w,h)=face
    #put the rectangle onto the test_img
    cv2.rectangle(test_img,(x,y),(x+w,y+h),(255,0,0),thickness=3)
```

```python
#put the text on the face rectangle 
def FaceRec_Function5(test_img,text,x,y):
#recieve the image,text the starting position of the image
    cv2.putText(test_img,text,(x,y),cv2.FONT_HERSHEY_COMPLEX,0.5,(0,0,255),2)
```

```python
video_capture=cv2.VideoCapture(0)
while True:
    _,frame=video_capture.read()
         
    faces_detected,grayimg=fn.FaceRec_Function1(frame) 
    #creating the label text for the subfolder faceid 
    name={0:"Anitha",1:"Anitha",2:"Dhayanithi"}       
    for face in faces_detected:
        (x,y,w,h)=face
        GrayImage=grayimg[y:y+w,x:x+h]
        label,confidence=face_recognition.predict(GrayImage)
        #draw the rectangle on the image
        FaceRec_Function4(frame,face)
        predicted_name=name[label]
        print(predicted_name)
        print(confidence)
        
        #checking confidence level to write name
        if confidence >150:
            continue
        else:
            FaceRec_Function5(frame,predicted_name,x,y)
            #resizedimage=cv2.resize(frame,(600,500))
    
    cv2.imshow('Video',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'): #Q  is the cancel key           
        break
 #release the video handle and destroy all windows
video_capture.release
cv2.destroyAllWindows()
            
```
